```python
import pandas as pd
import numpy as np
math_data = pd.read_csv("student-mat.csv")
```

```python
math_data.drop(['address', 'school', 'age', 'nursery', 'famsize', 'higher', 'failures', 'Mjob', 'romantic', 'Fjob', 'reason', 'goout', 'guardian', 'famsup', 'Dalc', 'Walc', 'G2'], axis = 1, inplace=True)
```

```python
data_dum = pd.get_dummies(math_data)
```

```python
data_dum.head()
```

```python
data_dum.drop(['Pstatus_A', 'schoolsup_no', 'paid_no', 'activities_no', 'internet_no', 'sex_F'], axis=1, inplace=True)
```

```python
data_dum.head()
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(data_dum, 
                        test_size=0.2, random_state=0)
print('Train size: ', len(train_set), 'Test size: ', len(test_set))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
X = train_set.drop('G3', axis=1)
y = train_set['G3']
X_test = test_set.drop('G3', axis=1)
y_test = test_set['G3']

lr_model = LinearRegression()
lr_model.fit(X,y)

y_pred = lr_model.predict(X)
print('Results for linear regression on training data')
print('Input: Independent Variables')
print(' Default settings')
print('Internal parameters:')
print(' Bias is ', lr_model.intercept_)
print(' Coefficients', lr_model.coef_)
print(' Score', lr_model.score(X,y))
print('MAE is ', mean_absolute_error(y, y_pred))
print('RMSE is ', np.sqrt(mean_squared_error(y, y_pred)))
print('MSE is ', mean_squared_error(y, y_pred))
print('R^2 ', r2_score(y,y_pred))

y_test_pred = lr_model.predict(X_test)
print()
print('Results for linear regression on test data')
print('Input: Independent Variables')
print('MAE is ', mean_absolute_error(y_test, y_test_pred))
print('RMSE is ', np.sqrt(mean_squared_error(y_test, 
y_test_pred)))
print('MSE is ', mean_squared_error(y_test, y_test_pred))
print('R^2 ', r2_score(y_test,y_test_pred))


```

```python
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
X = train_set.drop('G3', axis=1)
y = train_set['G3']
X_test = test_set.drop('G3', axis=1)
y_test = test_set['G3']

linmod_las = Lasso(alpha  = .73)
linmod_las.fit(X,y)

y_pred = linmod_las.predict(X)
print('Results for linear regression on training data')
print('Input: Independent Variables')
print(' Default settings')
print('Internal parameters:')
print(' Bias is ', linmod_las.intercept_)
print(' Coefficients', linmod_las.coef_)
print(' Score', linmod_las.score(X,y))
print('MAE is ', mean_absolute_error(y, y_pred))
print('RMSE is ', np.sqrt(mean_squared_error(y, y_pred)))
print('MSE is ', mean_squared_error(y, y_pred))
print('R^2 ', r2_score(y,y_pred))

y_test_pred = linmod_las.predict(X_test)
print()
print('Results for linear regression on test data')
print('Input: Independent Variables')
print('MAE is ', mean_absolute_error(y_test, y_test_pred))
print('RMSE is ', np.sqrt(mean_squared_error(y_test, 
y_test_pred)))
print('MSE is ', mean_squared_error(y_test, y_test_pred))
print('R^2 ', r2_score(y_test,y_test_pred))


```
