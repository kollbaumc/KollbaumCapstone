```python
#dropping the unwanted data attributes
data.drop(['address', 'school', 'age', 'nursery', 'famsize', 'higher', 'failures', 'Mjob', 'romantic', 'Fjob', 'reason', 'goout', 'guardian', 'famsup', 'Dalc', 'Walc', 'G1', 'G2'], axis = 1, inplace=True)
```

```python
data['New_Internet_Col'] = np.where(data['internet']=='no',0,1)
```

```python
data['New_Sex_Col'] = np.where(data['sex']=='F',0,1)
```

```python
data['New_Pstatus_Col'] = np.where(data['Pstatus']=='A',0,1)
```

```python
data['New_Activities_Col'] = np.where(data['activities']=='no',0,1)
```

```python
data['New_paid'] = np.where(data['paid']=='no',0,1)
```

```python
data['New_schoolsup'] = np.where(data['schoolsup']=='no',0,1)
```
